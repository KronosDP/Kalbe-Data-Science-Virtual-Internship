```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('1_Merged_Cleaned.csv')
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt

def plot_grouped_data(df, column, title):
    fig, ax = plt.subplots(2, 2, figsize=(12, 8))
    df.groupby("Day").mean()[column].plot(kind="line", ax=ax[0, 0])
    ax[0, 0].set_title("Mean {} by Day".format(title))
    ax[0, 0].grid(True)

    df.groupby("Month").mean()[column].plot(kind="line", ax=ax[0, 1])
    ax[0, 1].set_title("Mean {} by Month".format(title))
    ax[0, 1].grid(True)

    df.groupby("Day").sum()[column].plot(kind="line", ax=ax[1, 0])
    ax[1, 0].set_title("Sum of {} by Day".format(title))
    ax[1, 0].grid(True)

    df.groupby("Month").sum()[column].plot(kind="line", ax=ax[1, 1])
    ax[1, 1].set_title("Sum of {} by Month".format(title))
    ax[1, 1].grid(True)

    plt.tight_layout()
    plt.show()
```

```python
plot_grouped_data(df, "Qty", "Quantity")
plot_grouped_data(df, "TotalAmount", "Total Amount")
```
