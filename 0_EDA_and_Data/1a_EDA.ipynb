```python
df['Date'] = pd.to_datetime(df['Date'], format="%d/%m/%Y")  # Convert to datetime data type

df['Day'] = df['Date'].dt.day
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
```

```python
df.isna().sum()
```

```python
df[df["Marital Status"].isna()]
```

```python
# Umur rata rata yang gendernya pria udah menikah di umur berapa rata rata
df[(df["Gender"] == 1) & (df["Marital Status"] == "Married")]["Age"].mean()
```

```python
# Umur rata rata yang gendernya pria udah menikah di umur berapa minimal
df[(df["Gender"] == 1) & (df["Marital Status"] == "Married")]["Age"].min()
```

```python
graph = df[(df["Gender"] == 1) & (df["Marital Status"] == "Married")]["Age"]

plt.hist(graph, bins=30, edgecolor='black')  # Creating a histogram with 10 bins
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Distribution of Ages for Married Males")
plt.grid()
plt.show()
```

```python
# fill nan with married
df["Marital Status"] = df["Marital Status"].fillna("Married")
```

```python
df.loc[df["Age"] < 15, "Marital Status"] = "Single"
```
